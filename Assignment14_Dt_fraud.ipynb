```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
data = pd.read_csv('Fraud_check.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Converting Target variable 'Taxable Income' into categories risky and good.
data['Taxable.Income'] = pd.cut(x=data['Taxable.Income'],bins=[0,30000,100000], labels=['risky','good'], right = False)
data['Taxable.Income']
```

```python
data['Taxable.Income'].value_counts()
```

```python
#encoding categorical data
label_encoder = preprocessing.LabelEncoder()

data['Undergrad'] = label_encoder.fit_transform(data['Undergrad'])
data['Marital.Status'] = label_encoder.fit_transform(data['Marital.Status'])
data['Urban'] = label_encoder.fit_transform(data['Urban'])
data['Taxable.Income'] = label_encoder.fit_transform(data['Taxable.Income'])
data
```

```python
data.corr()
```

```python
data1 = data.drop('Taxable.Income', axis = 1)
data1
```

```python

# Dividing data into independent variables and dependent variable
x = data1.iloc[:,0:]
y = data['Taxable.Income']
```

```python
x
```

```python
y
```

```python
# Splitting data into training and testing data set
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.33, random_state=40)
```

```python
x_train
```

```python
x_test
```

```python

y_train
```

```python
y_test
```

```python
#Building Decision Tree Classifier using Entropy Criteria
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
#PLot the decision tree
plt.figure(figsize=(14,10))
tree.plot_tree(model);
```

```python
features=['under_grad', 'marital_status', 'city_population', 'work_experience',
       'urban', 'taxable_category']
opinion=['Risky', 'Good']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (6,6), dpi=600)
tree.plot_tree(model,
               feature_names = features, 
               class_names=opinion,
               filled = True);
```

```python
#Predicting on test data
pred = model.predict(x_test) # predicting on test data set 
pd.Series(pred).value_counts() # getting the count of each category 
```

```python
pred
```

```python
# getting the 2 way table to understand the correct and wrong predictions
pd.crosstab(y_test,pred)
```

```python
# Accuracy 
np.mean(pred==y_test)
```

```python
#Building Decision Tree Classifier (CART) using Gini Criteria
```

```python
model_gini = DecisionTreeClassifier(criterion='gini', max_depth=3)
model_gini
```

```python
model_gini.fit(x_train, y_train)
```

```python
# Plotting Decision tree
plt.figure(figsize=(14,10))
tree.plot_tree(model_gini);
```

```python
features=['under_grad', 'marital_status', 'city_population', 'work_experience',
       'urban', 'taxable_category']
opinion=['Risky', 'Good']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (6,6), dpi=600)
tree.plot_tree(model_gini,
               feature_names = features, 
               class_names=opinion,
               filled = True);
```

```python
# Predicting Data
pred = model_gini.predict(x_test)
pd.Series(pred).value_counts()
```

```python

pred
```

```python
# Creating cross tables for checking model
pd.crosstab(y_test, pred)
```

```python

# Checking accuracy of model
model_gini.score(x_test, y_test)
```
